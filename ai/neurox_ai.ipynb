```python
from google import genai

# Dummy API key for demo / structure
client = genai.Client(api_key="GEMINI_DUMMY_API_KEY_123456")
```

```python
def clarify_message(text):
    prompt = f"""
    Rewrite the message below to be:
    - Clear
    - Kind
    - Professional
    - Neurodiversity-friendly
    - Low stress

    Message:
    {text}
    """

    response = client.models.generate_content(
        model="gemini-1.5-flash",
        contents=prompt
    )

    return response.text


# Demo (structure only with dummy key)
clarify_message("I am overwhelmed and nobody responds on time")
```

```python
def burnout_check(hours_worked, tasks_done, breaks_taken):
    score = (hours_worked / (breaks_taken + 1)) - tasks_done

    if score >= 6:
        return "üö® High burnout risk detected. Strongly recommend rest."
    elif score >= 3:
        return "‚ö†Ô∏è Moderate workload. Consider a short break."
    else:
        return "‚úÖ Healthy, sustainable pace"


# Demo
burnout_check(9, 3, 1)
```
